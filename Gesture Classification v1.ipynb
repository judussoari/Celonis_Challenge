## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.interpolate as interp

%matplotlib inline
```

```python
userindex = 8
dayindex = 7
gestureindex = 8
repeatindex = 10
```

## User 1

```python
df = pd.DataFrame()
```

```python
#iteration over every .txt file for User 1
for j in range(1,dayindex+1):
    for k in range(1,gestureindex+1):
        for l in range(1,repeatindex+1):
            file = pd.read_csv("uWaveGestureLibrary/U1 ({})/A_Template_Acceleration{}-{}.txt".format(j,k,l),
                             header=None,sep=" ")
            #define the columns
            file.columns = ["X Acceleration","Y Acceleration","Z Acceleration"]

            vec1 = np.array([file["X Acceleration"],file["Y Acceleration"],file["Z Acceleration"]])

            vec21 = vec1[0] #X Acceleration
            vec22 = vec1[1] #Y Acceleration
            vec23 = vec1[2] #Z Acceleration

            vec31 = interp.interp1d(np.arange(vec21.size),vec21) #Interpolation of the respective values
            vec32 = interp.interp1d(np.arange(vec22.size),vec22)
            vec33 = interp.interp1d(np.arange(vec23.size),vec23)

            vec41 = vec31(np.linspace(0,vec21.size-1,100)) #To finally have a unified size for every time series
            vec42 = vec32(np.linspace(0,vec22.size-1,100))
            vec43 = vec33(np.linspace(0,vec23.size-1,100))

            vec = np.array([vec41,vec42,vec43]).reshape(300,) #1-row vector containing X,Y and Z
            vec = np.append(vec,k)
            ser = pd.Series(vec)
            df = pd.concat([df,ser],axis=1)
            #matrix = np.append(matrix,vec,axis=0)
```

## Multinomial Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
```

```python
df_transpo.head()
```

```python
#Task: Given a time series, perform gesture detection
X = df_transpo.drop("Label",axis=1)
y = df_transpo["Label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
#X_train = scaler.fit_transform(X_train)
#X_test = scaler.transform(X_test)
```

```python
#Multinomial, because I am dealing with 8 classes
#Solver: default one â€“ Others possible
#C: Regularization parameter. The closer to 0, the harder penalization
model = LogisticRegression(multi_class='multinomial', solver='lbfgs',penalty="l2",C=0.001)
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred[0:20]
```

## Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
#Large number of misclassifications
#Overall bad performance
#Options: Hyperparameter tuning, new ML model (SVM,ANN), Data Normalization, Cross-validation, 
confusion_matrix(y_test,pred)
```
