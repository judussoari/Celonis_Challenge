## User 1

```python
#emtpy DataFrame
u1 = pd.DataFrame(columns=["X Acceleration","Y Acceleration","Z Acceleration",
                           "User Index","Day Index","Repeat Index","Gesture Index"])

#iteration over every .txt file for User 1
for i in range(1,dayindex+1):
    for j in range(1,gestureindex+1):
        for k in range(1,repeatindex+1):
            df = pd.read_csv("uWaveGestureLibrary/U1 ({})/A_Template_Acceleration{}-{}.txt".format(i,j,k),
                             header=None,sep=" ")
            #define the columns
            df.columns = ["X Acceleration","Y Acceleration","Z Acceleration"]
            
            #add information that is stored in the .txt file names
            df["User Index"] = df["Z Acceleration"].apply(lambda x: 1)
            df["Day Index"] = df["Z Acceleration"].apply(lambda x: i)
            df["Repeat Index"] = df["Z Acceleration"].apply(lambda x: k) 
            df["Gesture Index"] = df["Z Acceleration"].apply(lambda x: j) 
            
            #after every iteration, concatenate the DataFrame from new file to previously concatenated df
            u1 = pd.concat([u1,df],ignore_index=True)
```

## Multinomial Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV, LogisticRegression
```

```python
df.head()
```

```python
#Task: Given a time series, perform gesture detection
X = df[["X Acceleration","Y Acceleration","Z Acceleration"]]
y = df["Gesture Index"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
#X_train = scaler.fit_transform(X_train)
#X_test = scaler.transform(X_test)
```

```python
#Multinomial, because I am dealing with 8 classes
#Solver: default one â€“ Others possible
#C: Regularization parameter. The closer to 0, the harder penalization
model = LogisticRegression(multi_class='multinomial', solver='lbfgs',penalty="l2",C=0.001)
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred[0:20]
```

## Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
#Large number of misclassifications
#Overall bad performance
#Options: Hyperparameter tuning, new ML model (SVM,ANN), Data Normalization, Cross-validation, 
confusion_matrix(y_test,pred)
```
